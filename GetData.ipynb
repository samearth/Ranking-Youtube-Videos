```python
comm = pd.read_csv('ALLcomments.csv',encoding='utf8',error_bad_lines=False);#opening the file UScomments
```

```python
comm = comm.drop(columns = ['video_id', 'likes', 'replies'])
```

```python
data = comm
```

```python
pol=[] # list which will contain the polarity of the comments
for i in data['comment_text']:
    try:
        analysis =TextBlob(i)
        pol.append(analysis.sentiment.polarity)
        
    except:
        pol.append(0)
```

```python
data['pol']=pol
```

```python
data['pol'][data['pol'] > 0] = 1
```

```python
data['pol'][data['pol'] < 0] = -1
```
